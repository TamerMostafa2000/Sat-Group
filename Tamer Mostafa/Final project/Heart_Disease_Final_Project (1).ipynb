```python
x=df.drop('HeartDisease_Yes',axis=1)
x
```

```python
y=df['HeartDisease_Yes']
y
```

```python
y.value_counts()
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=45,stratify=y)
y_train.value_counts()
```

```python
from imblearn.under_sampling import RandomUnderSampler
sampler=RandomUnderSampler()

x_train,y_train=sampler.fit_resample(x_train,y_train)
```

```python
y_train.value_counts()
```

```python
x_train['BMI']
```

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
x_train[['BMI']]=scaler.fit_transform(x_train[['BMI']])
x_test[['BMI']]=scaler.transform(x_test[['BMI']])
x_test
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.ensemble import RandomForestClassifier
from xgboost import XGBClassifier
from sklearn.svm import SVC

from sklearn.metrics import accuracy_score
```

```python
models={
    'lr':LogisticRegression(),
    'KNN':KNeighborsClassifier(),
    'DT':DecisionTreeClassifier(),
    'SVC':SVC(),
    'RF':RandomForestClassifier(),
    'XGBC':XGBClassifier(),
    'NB':GaussianNB()
    
}
```

```python
for name,model in models.items():
    print(f'using:{name}')
    model.fit(x_train,y_train)
    print(f'training accuracy: {model.score(x_train,y_train)}')
    print(f'testing accuracy: {model.score(x_test,y_test)}')
    print('_'*30)
```

```python
#Data for poeple with heart diseases.
df=pd.read_csv('/content/heart_2020_cleaned.csv')
#Removing of duplicates
df.drop_duplicates
#converting (sleepTime) from a numeric to a Categorical feature (Sleeping_Period)
def sleeping_p(x):
  if x < 5 :
    return 'Less than 5'
  if x ==5 :
    return '5'
  if x ==6  :
    return '6'
  if x ==7 :
    return '7' 
  if x ==8 :
    return '8'
  if x ==9 :
    return '9' 
  if x ==10 :
    return '10'
  if x > 10 :
    return 'More than 10'  
df['Sleeping_Period']=df['SleepTime'].apply(sleeping_p)
df.drop('SleepTime',axis=1,inplace=True)

#converting (PhysicalHealth) from a numeric to a Categorical feature.
def phy_Health_cat(x):
  if x ==0 :
    return 'Normal'
  if x >=1 and x <= 5:
    return 'Mild degree'
  if x >5 and x <=10 :
    return 'Moderate degree'
  if x >10 and x <=20 :
    return 'Marked degree'
  if x >20 :
    return 'Severe degree' 
df['PhysicalHealth']=df['PhysicalHealth'].apply(lambda x: phy_Health_cat(x))

#converting (mentalHealth) from a numeric to a Categorical feature(MentalHealth (0 to 4)) 0 is normal, 4 is very bad.

def Mental_Health_cat(x):
  if x ==0 :
    return '0'
  if x >=1 and x <= 5:
    return '1'
  if x >5 and x <=10 :
    return '2'
  if x >10 and x <=20 :
    return '3'
  if x >20 and x <=30 : 
    return '4' 
df['MentalHealth']=df['MentalHealth'].apply(lambda x: Mental_Health_cat(x))


#deleting BMI outlires
from datasist.structdata import detect_outliers
outlier_indices=detect_outliers(df,0,['BMI'])
outlier_indices
df.drop(outlier_indices,inplace=True)
```
